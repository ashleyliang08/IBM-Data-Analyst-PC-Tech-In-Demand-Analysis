### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
!pip install seaborn
import seaborn as sns
```

```python
##Write your code here
df_age_freq = df[df['SOPartFreq'].notna()]
age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
freq_order = [
    'I have never participated in Q&A on Stack Overflow',
    'Less than once per month or monthly',
    'A few times per month or weekly',
    'A few times per week',
    'Daily or almost daily',
    'Multiple times per day'

]

age_mapping = {age: idx1 for idx1, age in enumerate(age_order)}
freq_mapping = {freq: idx2 for idx2, freq in enumerate(freq_order)}
#
df_age_freq['Age_numeric'] = df_age_freq['Age'].map(age_mapping)
df_age_freq['SOPartFreq_numeric'] = df_age_freq['SOPartFreq'].map(freq_mapping)

#plot
plt.figure(figsize=(8,6))
sns.scatterplot(data = df_age_freq,x = 'Age_numeric',y = 'SOPartFreq_numeric', size = 'JobSat',sizes=(20, 100))
plt.xticks(
    ticks=range(len(age_order)),
    labels=age_order,
    rotation=45)
plt.yticks(
    ticks=range(len(freq_order)),
    labels=freq_order)

plt.legend(title='Job Satisfaction', bbox_to_anchor=(1, 1),
           loc='upper left',markerscale=1.15,borderpad=0.5,frameon = False)
plt.show()





```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python

# 假设df是已加载的原始数据
# 1. 数据预处理：拆分语言列并展开为多行
# 先处理缺失值，然后拆分语言字符串
df_lang_expanded = df.dropna(subset=['LanguageHaveWorkedWith', 'Age']).copy()
df_lang_expanded['Language'] = df_lang_expanded['LanguageHaveWorkedWith'].str.split(';')
df_lang_expanded = df_lang_expanded.explode('Language')  # 每行只保留一种语言

# 2. 获取最受欢迎的前5种语言
top_5_lang = df_lang_expanded['Language'].value_counts().head(5).index.tolist()

# 3. 筛选出前5种语言的数据，并按年龄和语言分组计算频率
df_age_lang = df_lang_expanded[df_lang_expanded['Language'].isin(top_5_lang)]
lang_age_counts = df_age_lang.groupby(['Age', 'Language']).size().reset_index(name='Frequency')

# 4. 设置年龄顺序（确保绘图时按正确顺序显示）
age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older',
]
lang_age_counts['Age'] = pd.Categorical(lang_age_counts['Age'], categories=age_order, ordered=True)

# 5. 绘制气泡图
plt.figure(figsize=(14, 8))
scatter = sns.scatterplot(
    data=lang_age_counts,
    x='Language',          # x轴：编程语言
    y='Age',               # y轴：年龄段
    size='Frequency',      # 气泡大小：频率（反映popularity）
    hue='Age',             # 按年龄着色
    sizes=(50, 500),       # 气泡大小范围
    alpha=0.7,             # 透明度
    palette='viridis'      # 颜色方案
)

# 6. 美化图例和图表
plt.title('Popularity of Programming Languages by Age Group', fontsize=18, pad=20)
plt.xlabel('Programming Languages', fontsize=14, labelpad=10)
plt.ylabel('Age Groups', fontsize=14, labelpad=10)

# 调整图例
handles, labels = scatter.get_legend_handles_labels()
# 分离大小图例和颜色图例
size_idx = labels.index('Frequency')  # 找到大小图例的位置
scatter.legend(
    handles[:size_idx+1] + handles[size_idx+2:],  # 排除重复的年龄标题
    labels[:size_idx+1] + labels[size_idx+2:],
    title='',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    fontsize=11
)

plt.tight_layout()
plt.show()
```
