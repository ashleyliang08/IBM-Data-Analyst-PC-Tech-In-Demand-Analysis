
### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




法一：用sns

```python
## Write your code here

age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
df['Age'] = pd.Categorical(df['Age'], categories=age_order, ordered=True)

plt.figure(figsize = (8,5))
sns.scatterplot(data = df,x = 'Age',y ='JobSatPoints_6')
plt.title('Scatter Plot for Age vs. Job Satisfaction')
plt.xticks(rotation = 10)
plt.ylabel('Job Satisfaction')
plt.xlabel('Age')
plt.show()
```

法二：用matplotlib


```python
# 将年龄列转换为有序分类类型
age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]

# 创建年龄到数值的映射字典
age_mapping = {age: idx for idx, age in enumerate(age_order)}

# 添加数值列用于绘图
df['Age_numeric'] = df['Age'].map(age_mapping)
df['Age_numeric']
plt.figure(figsize=(8, 5))
plt.scatter(data=df, x='Age_numeric', y='JobSatPoints_6')
plt.title('Scatter Plot for Age vs. Job Satisfaction')
# 设置x轴刻度和标签
plt.xticks(ticks=range(len(age_order)),
           labels=age_order,
           rotation=10)
plt.ylabel('Job Satisfaction')
plt.xlabel('Age')
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
plt.figure(figsize=(10, 6))
sns.scatterplot(data =df,x = 'ConvertedCompYearly',y = 'JobSatPoints_6')
plt.title('Scatter Plot for Compensation vs. Job Satisfaction')
# 设置x轴刻度和标签
plt.ylabel('Job Satisfaction')
plt.xlabel('Compensation')
plt.xlim(0,900000)
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here

age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
age_mapping = {age: idx for idx, age in enumerate(age_order)}

# 添加数值列用于绘图
df['Age_numeric'] = df['Age'].map(age_mapping)

plt.figure(figsize = (8,5))
sns.regplot(data = df,x = 'Age_numeric',y ='JobSatPoints_6')
plt.title('Scatter Plot for Age vs. Job Satisfaction')
plt.xticks(ticks=range(len(age_order)),
           labels=age_order,rotation = 10)
plt.ylabel('Job Satisfaction')
plt.xlabel('Age')
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
## Write your code here
def prof_coding_years(years):
    if years == 'Less than 1 year':
        return 0.5
    elif years == 'More than 50 years':
        return 50
    else:
        try:
            return float(years)
        except:
            return None



df['YearsCodePro_num'] = df['YearsCodePro'].apply(prof_coding_years)

plt.figure(figsize = (9,5))
sns.regplot(data = df,x = 'Age_numeric',y ='YearsCodePro_num')
plt.title('Scatter Plot for Age vs. Job Satisfaction')
plt.xticks(ticks=range(len(age_order)),
           labels=age_order,rotation = 10)
plt.ylabel('Years of Coding Profetionally')
plt.ylim(0,60)
plt.xlabel('Age')
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here

top_5_emp = list(df['Employment'].value_counts().head(5).index)
df_top5 = df[df['Employment'].isin(top_5_emp)]

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_top5, x='ConvertedCompYearly', y='JobSatPoints_6',
                hue='Employment', style='Employment',size = 'Employment')

plt.title('Scatter Plot for Compensation vs. Job Satisfaction (Top 5 Employment Types)',fontsize = 15,loc = 'center')
plt.ylabel('Job Satisfaction',fontsize = 13)
plt.xlabel('Compensation',fontsize = 13)
plt.xlim(0, 900000)
plt.legend(title='Employment Type', bbox_to_anchor=(0.5, 1),
           loc='upper left',markerscale=1.15,borderpad=0.5,frameon = False)
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
top_5_coun = list(df['Country'].value_counts().head(5).index)
df_top5_coun = df[df['Country'].isin(top_5_coun)]

plt.figure(figsize=(12, 6))
sns.scatterplot(data = df_top5_coun, x='Age', y='YearsCodePro_num',
                hue='Country', style='Country',size = 'Country')

plt.title('Scatter Plot for Work Experience vs. Age Group by Top 5 Country',fontsize = 15,loc = 'center')
plt.ylabel('Work Experience(Years) ',fontsize = 13)
plt.xlabel('Age',fontsize = 13)
plt.legend(title='Country', bbox_to_anchor=(0.5, 1),
           loc='upper right',markerscale=1.15,borderpad=0.5,frameon = False)
plt.show()
```
